```python
import pandas as pd
import time
%matplotlib inline

userCSV = pd.read_csv('Users/ec6dfcf19485cb011e0b22637075037aae34cf26_FV.csv')
```

```python
artistNames = userCSV.artistName.tolist()

```

```python
from wordcloud import WordCloud, STOPWORDS


text= ', '.join(str(e) for e in artistNames) #.decode("utf-8")


wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color='white',
                          width=1200,
                          height=1000).generate(text)


import matplotlib.pyplot as plt

fig = plt.gcf()

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
plt.draw()
fig.savefig('Plots/user_artist2_ec6dfcf19485cb011e0b22637075037aae34cf26.png',dpi=100)

plt.clf()
plt.cla()
plt.close()

# lower max_font_size
wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color='white',
                          width=1200,
                          height=1000,
                          max_font_size=250).generate(text)

fig = plt.gcf()

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
plt.draw()
fig.savefig('Plots/user_artist1_ec6dfcf19485cb011e0b22637075037aae34cf26.png',dpi=100)

plt.clf()
plt.cla()
plt.close()

```
